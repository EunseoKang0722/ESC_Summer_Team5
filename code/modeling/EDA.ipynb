```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('cluster.csv')
```

```python
data.head()
```

## EDA

```python
hit0 = data[data['cluster_team_hit']==0]
hit1 = data[data['cluster_team_hit']==1]
pit0 = data[data['cluster_team_pit']==1]
pit1 = data[data['cluster_team_pit']==2]
```

```python
fig, ax = plt.subplots(1, 3, figsize=(10, 5))
ax[0].boxplot([hit0['hit_f1'], hit1['hit_f1']])
ax[0].set_xticklabels(['cluster 1', 'cluster 2'])
ax[0].set_xlabel('HIT F1')
ax[1].boxplot([hit0['hit_f2'], hit1['hit_f2']])
ax[1].set_xticklabels(['cluster 1', 'cluster 2'])
ax[1].set_xlabel('HIT F2')
ax[2].boxplot([hit0['hit_f3'], hit1['hit_f3']])
ax[2].set_xticklabels(['cluster 1', 'cluster 2'])
ax[2].set_xlabel('HIT F3')
plt.show()
```

```python
from mpl_toolkits import mplot3d
```

```python
fig = plt.figure(figsize=(8, 8))
ax = plt.axes(projection='3d')
ax.scatter3D(data['hit_f1'], data['hit_f2'], data['hit_f3'], c=data['cluster_team_hit'])
ax.set_xlabel('HIT F1')
ax.set_ylabel('HIT F2')
ax.set_zlabel('HIT F3')
plt.show()
```

```python
fig, ax = plt.subplots(1, 4, figsize=(14, 5))
ax[0].boxplot([pit0['pit_f1'], pit1['pit_f1']])
ax[0].set_xticklabels(['cluster 1', 'cluster 2'])
ax[0].set_xlabel('PIT F1')
ax[1].boxplot([pit0['pit_f2'], pit1['pit_f2']])
ax[1].set_xticklabels(['cluster 1', 'cluster 2'])
ax[1].set_xlabel('PIT F2')
ax[2].boxplot([pit0['pit_f3'], pit1['pit_f3']])
ax[2].set_xticklabels(['cluster 1', 'cluster 2'])
ax[2].set_xlabel('PIT F3')
ax[3].boxplot([pit0['pit_f4'], pit1['pit_f4']])
ax[3].set_xticklabels(['cluster 1', 'cluster 2'])
ax[3].set_xlabel('PIT F4')
plt.show()
```
