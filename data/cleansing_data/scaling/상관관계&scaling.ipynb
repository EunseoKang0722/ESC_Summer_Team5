### H/9 & OAVG  상관관계가 매우 높음 -> 둘 다 얼마나 안타를 맞았는지에 관한 지표
### HR/9 & kFIP 상관관계 높음

### RA9 & ERA & OOPS 상관관계 높음

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

## 투수 scaling OVAG ~ OOPS, BABIP를 제외한 나머지 값들을 다 0~1 사이로 scaling 함
col = ['K/9','BB/9','K/BB','WHIP','DICE','ERA','RA9','FIP','kFIP','HR/9','H/9']
pitcher_scaler_2016 = scaler.fit_transform(pitcher_data_2016.loc[:,col])
pitcher_scaler_2016 = pd.DataFrame(pitcher_scaler_2016)
pitcher_scaler_2016.columns = col
pitcher_data_2016[col] = pitcher_scaler_2016

pitcher_scaler_2017 = scaler.fit_transform(pitcher_data_2017.loc[:,col])
pitcher_scaler_2017 = pd.DataFrame(pitcher_scaler_2017)
pitcher_scaler_2017.columns = col
pitcher_data_2017[col] = pitcher_scaler_2017

pitcher_scaler_2018 = scaler.fit_transform(pitcher_data_2018.loc[:,col])
pitcher_scaler_2018 = pd.DataFrame(pitcher_scaler_2018)
pitcher_scaler_2018.columns = col
pitcher_data_2018[col] = pitcher_scaler_2018

pitcher_scaler_2019 = scaler.fit_transform(pitcher_data_2019.loc[:,col])
pitcher_scaler_2019 = pd.DataFrame(pitcher_scaler_2019)
pitcher_scaler_2019.columns = col
pitcher_data_2019[col] = pitcher_scaler_2019

pitcher_scaler_2020 = scaler.fit_transform(pitcher_data_2020.loc[:,col])
pitcher_scaler_2020 = pd.DataFrame(pitcher_scaler_2020)
pitcher_scaler_2020.columns = col
pitcher_data_2020[col] = pitcher_scaler_2020
```

### RC & XR -> 둘 다 득점 생산력 의미 (상관관계 매우 높음) => 1개만 써도 무방할듯
### wOBA & OPS이 상관관계 높음(0.960) -> 제거여부 고민 필요
* OPS(출루율 + 장타율) -> 가장 인정받는 타자 지표라고 함
* wOBA(가중출루율) -> OPS와 비슷하나 분석결과와 평가방법이 다르다고 함

### wOBA & OBP 상관관계 높음 -> wOBA는 OBP의 분자에 들어가는 값에 다른 가중치를 적용하였기 때문에
* -> wOBA만 써도 될듯

### OPS& SLG 상관관계가 높음(0.973) -> OPS가 SLG 포함하기 때문에 -> OPS만 써도 될듯
* OPS가 실제 공격 기여 수준과 높은 상관관계를 갖고 있다고 함

### IsOP & SLG 상관관계가 높음 -> IsOP 계산식에 SLG이 포함되기 때문
### 년도별로 바라보니 AVG&OBP 상관관계 높음(0.9정도)
### RUN(득점)과 상관관계가 높았던 것들 -> 'RC', 'XR', 'P_HRA_RT(득점권 타율)'

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()

## 타자에서는 XR, RC를 제외하고는 다 0~1사이이기에 이 둘을 0~!사이로 scaling하는 선택을 함 (시즌별로)
hitter_scaler_2016 = scaler.fit_transform(hitter_data_2016.loc[:,['RC','XR']])
hitter_scaler_2016 = pd.DataFrame(hitter_scaler_2016)
hitter_scaler_2016.columns = ['RC','XR']
hitter_data_2016[['RC','XR']] = hitter_scaler_2016

hitter_scaler_2017 = scaler.fit_transform(hitter_data_2017.loc[:,['RC','XR']])
hitter_scaler_2017 = pd.DataFrame(hitter_scaler_2017)
hitter_scaler_2017.columns = ['RC','XR']
hitter_data_2017[['RC','XR']] = hitter_scaler_2017

hitter_scaler_2018 = scaler.fit_transform(hitter_data_2018.loc[:,['RC','XR']])
hitter_scaler_2018 = pd.DataFrame(hitter_scaler_2018)
hitter_scaler_2018.columns = ['RC','XR']
hitter_data_2018[['RC','XR']] = hitter_scaler_2018

hitter_scaler_2019 = scaler.fit_transform(hitter_data_2019.loc[:,['RC','XR']])
hitter_scaler_2019 = pd.DataFrame(hitter_scaler_2019)
hitter_scaler_2019.columns = ['RC','XR']
hitter_data_2019[['RC','XR']] = hitter_scaler_2019

hitter_scaler_2020 = scaler.fit_transform(hitter_data_2020.loc[:,['RC','XR']])
hitter_scaler_2020 = pd.DataFrame(hitter_scaler_2020)
hitter_scaler_2020.columns = ['RC','XR']
hitter_data_2020[['RC','XR']] = hitter_scaler_2020
```
