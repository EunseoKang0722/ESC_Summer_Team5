# RC = 득점기여도

* RC = (A x B) / C
* A = 안타 + 볼넷 + 사구 - 도루실패 - 병살타
* B = 루타수 + 0.52 x (도루 + 희타 + 희생플라이) + 0.26 x (볼넷 + 사구 - 고의4구)
* C = 타수 + 볼넷 + 사구 + 희타 +  희생플라이

```python
d0 = data_final[['4구']]
d0_ = d0.values
d1 = data_final[['고4']]
d1_ = d1.values
d2 = data_final[['사구']]
d2_ = d2.values
d3 = data_final[['도루실패']]
d3_ = d3.values
d4 = data_final[['병살타']]
d4_ = d4.values
```

```python
A_ = b_ + d0_ + d2_ - d3_ - d4_
A_
```

```python
d5 = data_final[['희타']]
d5_ = d5.values
d6 = data_final[['희비']]
d6_ = d6.values
d7 = data_final[['도루']]
d7_ = d7.values
d8 = data_final[['삼진']]
d8_ = d8.values
```

```python
B_ = c4_ + 0.52*(d7_+d5_+d6_) + 0.26*(d0_+d2_-d1_)
```

```python
C_ = a_ + d0_ + d2_ + d5_ + d6_
```

```python
# RC 계산
x3_ = (A_*B_)/C_
x3_
```

```python
for i in range(len(x3_)):
    if C_[i]==0:
        x3_[i]=0
```

```python
x3=pd.DataFrame(x3_)
x3.columns=['RC']
x3
```

# wRC = wOBA기반 득점생산

# wRAA = 리그평균 대비 득점생산
* 타석 * (wOBA - 리그 wOBA) / (wOBA scale) 

```python
# wRAA 계산
w1_ = (((y1_ - 0.362)/1.097) * a2_[0:378,])
w2_ = (((y2_ - 0.351)/1.107) * a2_[378:738,])
w3_ = (((y3_ - 0.351)/1.075) * a2_[738:1069,])
w4_ = (((y4_ - 0.335)/1.242) * a2_[1069:1422,])
w5_ = (((y5_ - 0.346)/1.14)  * a2_[1422:,])

x5_ = np.append(w1_,w2_)
x5_ = np.append(x5_,w3_)
x5_ = np.append(x5_,w4_)
x5_ = np.append(x5_,w5_)

x5 = pd.DataFrame(x5_)
x5.columns=['wRAA']
x5
```
