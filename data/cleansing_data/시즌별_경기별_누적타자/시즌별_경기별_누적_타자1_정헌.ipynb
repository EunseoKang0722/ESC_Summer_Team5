# 타율 = 안타/타수

```python
a = data1[['타수']]
b = data1[['안타']]
a_ = a.values 
b_ = b.values

# 타율 계산
x_ = b_/a_
```

```python
# NaN 0으로 바꿈
for i in range(len(b_)):
    if b_[i] == 0:
        x_[i] = 0  
        
x = pd.DataFrame(x_)
x.columns=['BA']
```

# 장타율 = (1루타 * 1 + 2루타 * 2 + 3루타 * 3 + 홈런 * 4) / 타수

```python
c1 = data1[['2루타']]
c2 = data1[['3루타']]
c3 = data1[['홈런']]

c1_ = c1.values
c2_ = c2.values
c3_ = c3.values

# 1루타 계산
c0_ = b_ - (c1_+c2_+c3_)

# 루타수 계산
c4_ = (c0_*1) + (c1_*2) + (c2_*3) + (c3_*4)

# 장타율 계산
x1_ = c4_ / a_
```

```python
# NaN 0으로 바꿈
for i in range(len(x1_)):
    if a_[i]==0:
        x1_[i]=0
        
x1=pd.DataFrame(x1_)
x1.columns = ['SLG']
```

# 절대 장타율 = 장타율 - 타율

```python
# 절대 장타율 계산
x2_ = x1_ - x_
x2 = pd.DataFrame(x2_)
x2.columns=['IsoP']
```

# RC = 득점기여도

* RC = (A x B) / C
* A = 안타 + 볼넷 + 사구 - 도루실패 - 병살타
* B = 루타수 + 0.52 x (도루 + 희타 + 희생플라이) + 0.26 x (볼넷 + 사구 - 고의4구)
* C = 타수 + 볼넷 + 사구 + 희타 +  희생플라이

```python
d0 = data1[['4구']]
d1 = data1[['고4']]
d2 = data1[['사구']]
d3 = data1[['도루실패']]
d4 = data1[['병살타']]
d5 = data1[['희타']]
d6 = data1[['희비']]
d7 = data1[['도루']]
d8 = data1[['삼진']]
d0_ = d0.values
d1_ = d1.values
d2_ = d2.values
d3_ = d3.values
d4_ = d4.values
d5_ = d5.values
d6_ = d6.values
d7_ = d7.values
d8_ = d8.values
```

```python
A_ = b_ + d0_ + d2_ - d3_ - d4_
B_ = c4_ + 0.52*(d7_+d5_+d6_) + 0.26*(d0_+d2_-d1_)
C_ = a_ + d0_ + d2_ + d5_ + d6_
```

```python
# RC 계산 -> NaN 그대로 놔둠
x3_ = (A_*B_)/C_

x3=pd.DataFrame(x3_)
x3.columns=['RC']
```

# XR
 * 추정득점 -> 타자의 득점공헌도 (RC보다 좋다)
 * 1루타  x 0.5 + 2루타 x 0.72 + 3루타 x 1.04 + 홈런 x 1.44 + (사구+볼넷-고의사구) x 0.34 + 고의사구 x 0.25 + 도루 x 0.18 - 도루실패 x 0.32 - (타수-안타-삼진) x 0.09 - 삼진 x 0.098 - 병살타 x 0.37 + 희생플라이 x 0.37 + 희생번트 x 0.04

```python
# XR 계산
xr_ = c0_ * 0.5 + c1_ * 0.72 + c2_ * 1.04 + c3_ * 1.44 + (d2_+d0_-d1_) * 0.34 + d1_ * 0.25 + d7_ * 0.18 - d3_ * 0.32 - (a_ - b_ - d8_) * 0.09 - d8_ * 0.098 - d4_ * 0.37 + d6_ * 0.37 + d5_ * 0.04

xr = pd.DataFrame(xr_)
xr.columns = ['XR']
```
