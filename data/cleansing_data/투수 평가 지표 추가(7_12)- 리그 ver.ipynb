```python
import numpy as np
import pandas as pd
```

```python
pitcher= pd.read_csv('개인투수.csv',encoding= "utf-8")
pitcher
```

```python
pitcher_grouped= pitcher.drop(['Unnamed: 0','G_ID','GDAY_DS','T_ID','VS_T_ID','TB_SC','WLS','P_WHIP_RT',
                         'P2_WHIP_RT','CB_WHIP_RT','P_ID'], axis=1)

pitcher_grouped= pitcher_grouped.groupby(['NAME', 'GYEAR'], as_index=False).mean()

pitcher_grouped
```

```python
pitcher_grouped['ERA']= 9*pitcher_grouped['ER']/(pitcher_grouped['INN2']/3)
pitcher_grouped.loc[pitcher_grouped['INN2']==0, 'ERA']=np.nan #INN2가 0인 행은 모두 nan처리
pitcher_grouped.loc[pitcher_grouped['INN2']==0, 'ERA']
```

```python
pitcher_grouped['C']= pitcher_grouped['ERA'] - (13 * pitcher_grouped['HR']  + 3 *(pitcher_grouped['BB'] +pitcher_grouped['HP']) -pitcher_grouped['KK']) / (pitcher_grouped['INN2']/3)
C= pitcher_grouped[['NAME','GYEAR','C']]
```

```python
pitcher_grouped['C'].mean()
```

```python
pd.merge(pitcher,C)
```

```python
#7. DIPS의 개량된 형태 중 하나인 'DICE' 열 추가
pitcher['DICE']= 3.00 + (13* pitcher['HR'] + 3*( pitcher['BB']+ pitcher['HP'])-2* pitcher['KK'])/( pitcher['INN2'] /3 )
pitcher
```

```python
pitcher['ERA'] = 9 * pitcher['ER'] / (pitcher['INN2']/ 3) 
pitcher.head()
```

```python
#8-1 FIP 추가 (계수 C 계산 포함)
C= pitcher['ERA']/len() - (13 * pitcher['HR'] + 3 *(pitcher['BB'] + pitcher['HP']) - 2 * pitcher['KK'] ) / (pitcher['INN2']/3)
C #보통 3.1~3.2값이 나온다고 하는데,,, 이상하다.
```

```python
#8-2. '한국프로야구에서 FIP 계수의 추정' 에서 제시한 kFIP 추가
pitcher['kFIP'] = C + (14* pitcher['HR'] + 3*(pitcher['BB']+pitcher['HP']) - pitcher['KK'])/ (pitcher['INN2']/3)
```

```python
#9. RA9 (Run Allowed Per 9 Innings Pitched; 9이닝당 평균실점) 추가
pitcher['RA9'] = 9 * pitcher['R'] / (pitcher['INN2'] / 3)
```

```python
#10. HR/9 (Home Runs Allowed Per 9 Innings Pitched;9이닝당 피홈런 개수) 추가
pitcher['HR/9'] = 9 * pitcher['HR'] / (pitcher['INN2'] / 3)
```

```python
#11. H/9 (Hits Runs Allowed Per 9 Innings Pitched; 9이닝당 피안타 개수)
pitcher['H/9'] = 9 * pitcher['HIT'] / (pitcher['INN2'] / 3)
```

```python
#OAVG/OOBP/OSLG/OOPS – 피안타율/피출루율/피장타율/피OPS
pitcher['OAVG'] = pitcher['HIT']/ pitcher['AB']
pitcher['OOBP'] = (pitcher['HIT'] + pitcher['BB'] + pitcher['HP'] ) / (pitcher['AB'] + pitcher['BB'] + pitcher['HP'] + pitcher['SF'])
pitcher['OSLG'] = ( pitcher['HIT'] + 2*pitcher['H2'] + 3*pitcher['H3']+ 4*pitcher['HR'])/pitcher['AB']
pitcher['OOPS'] = pitcher['OOBP']+ pitcher['OSLG']
```
