### 타자지표 distplot 두팀 골라서 사용

```python
sns.set(font='HYGothic-Medium', font_scale=1.3, style='whitegrid')
team=thit['T_ID'].unique( )
columns=['PA', 'AB',  'RBI', 'RUN', 'HIT', 'H2', 'H3', 'HR', 'SB', 'CS', 'SH', 'SF', 'BB',
       'IB', 'HP', 'KK', 'GD', 'ERR', 'LOB', 'P_HRA_RT', 'P_AB_CN', 'P_HIT_CN']
@interact(team1 = team, team2=team, columns=columns)

def update(team1, team2, columns):
    fig, ax = plt.subplots()
    team1_data=thit19.loc[thit19['T_ID']==team1]
    sns.distplot(team1_data[columns], kde=False,label=team1) 
    team2_data=thit19.loc[thit19['T_ID']==team2]
    sns.distplot(team2_data[columns], kde=False,label=team2)
    ax.set_title('Distribution Plot')
```

### 타자지표 시도표

```python
sns.set(font='HYGothic-Medium', font_scale=1.3, style='whitegrid')
year=[2016,2017,2018,2019]

team=thit['T_ID'].unique( )

columns=['PA', 'AB',  'RBI', 'RUN', 'HIT', 'H2', 'H3', 'HR', 'SB', 'CS', 'SH', 'SF', 'BB',
       'IB', 'HP', 'KK', 'GD', 'ERR', 'LOB', 'P_HRA_RT', 'P_AB_CN', 'P_HIT_CN']

@interact(team = team, year=year,columns=columns)
def update(team, year, columns):
    fig, ax = plt.subplots(figsize=(15,5))
    year_data= thit.loc[thit['GYEAR']==year]
    team_data=year_data.loc[year_data['T_ID']==team]
    sns.lineplot(x='GDAY_DS', y=columns, data=team_data, ci=None)
    ax.set_title(str(year)+'시즌 '+str(team)+'팀 '+str(columns)+' 시도표')
```
